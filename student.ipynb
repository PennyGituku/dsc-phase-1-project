```python
# Selecting the relevant columns for the correlation analysis
selected_columns = ['domestic_gross', 'foreign_gross', 'production_budget', 'worldwide_gross', 'profit']

# Creating a new DataFrame containing only the selected columns
selected_df = df[selected_columns]

# Compute the correlation matrix
correlation_matrix = selected_df.corr()

# Unstacking the correlation matrix and resetting index to create a new DataFrame
ft_corr = correlation_matrix.unstack().reset_index()
ft_corr.columns = ['Attribute 1', 'Attribute 2', 'Correlation']

# Filtering out the self-correlations and duplicates
ft_corr = ft_corr[ft_corr['Attribute 1'] != ft_corr['Attribute 2']]
ft_corr.drop_duplicates(subset='Correlation', inplace=True)

# Sorting values in descending order
ft_corr.sort_values('Correlation', ascending=False, inplace=True)

# Creating a pivot table to reshape the DataFrame for heatmap plotting
pivot_table = ft_corr.pivot_table(index='Attribute 1', columns='Attribute 2', values='Correlation')

# Plotting the heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)

# Rotating the y-axis labels to be horizontal
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

# Rotating the x-axis label (Title 1) to be horizontal
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

# Rotating the y-axis label (Attribute 1) to be horizontal
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.title("Correlation Heatmap of Attributes with Gross Profit")
plt.show()
```

The heatmap that we plotted shows the correlation matrix of the selected attributes with respect to gross profit.
The heatmap is color-coded to represent the correlation coefficients going from cool (blue) to warm (red) to help visualize the strength and direction of the correlations:

Dark blue (cool) represents negative correlations, meaning that as one attribute increases, the other decreases.
Dark red (warm) represents positive correlations, meaning that as one attribute increases, the other also increases.
Lighter colors indicate weaker correlations.

We can observe a Positive Correlation between:
1. 'production_budget' and 'worldwide_gross'.
2. 'production_budget' and 'profit'.
3. 'worldwide_gross' and 'profit' have the highest correlation.

All the rest have negative correlations.
This heatmap helps us identify which attributes have stronger relationships. 'production_budget' and 'profit' have a positive correlation, indicating that movies with higher production budgets tend to have higher profits, but this correlation is not as strong as the one observed between 'profit' and 'worldwide_gross'.
